```python
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'],errors = 'coerce')
```

```python
# Visualize the distribution of key numerical variables to understand customer behavior
num_cols = ['tenure', 'MonthlyCharges','TotalCharges']
for col in num_cols:
    plt.figure()
    sns.histplot(data[col], kde = True) 
```

```python
churn_counts = data['Churn'].value_counts(normalize=True)*100
print(churn_counts)
```

```python
bins = [0,12,36,data['tenure'].max()]
labels = ['0-12 months', '13-36 months', '37+ months']

data['TenureGroup'] = pd.cut(data['tenure'],bins = bins, labels=labels, include_lowest=True)
```

```python
tenure_counts = data['TenureGroup'].value_counts()
print(tenure_counts)

```

```python
plt.figure(figsize = (6,6))
plt.pie(tenure_counts,labels= tenure_counts.index,autopct='%1.1f%%', startangle=90)
plt.title('Customer Distribution By Tenure')
plt.show()
```

```python
avg_charges = data.groupby('TenureGroup')['MonthlyCharges'].mean().reset_index()
print(avg_charges)
```

```python
plt.figure(figsize=(7,5))
bars = plt.bar(avg_charges['TenureGroup'], avg_charges['MonthlyCharges'])
plt.title('Average Monthly Charges Across Tenure Categories')
plt.xlabel('Tenure category')
plt.ylabel('Average monthly charges')

#add annotation
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.1f}',xy=(bar.get_x()+bar.get_width()/2, height),xytext=(0,3),textcoords='offset points', ha = 'center', va = 'bottom')
```

```python
bins = [0,12,36,60,data['tenure'].max()]
labels = ['0-12 months', '13-36 months', '37+ months','61+ months']

data['TenureGroup'] = pd.cut(data['tenure'],bins = bins, labels=labels, include_lowest=True)
```

```python
charge_stats = data.groupby('TenureGroup')['MonthlyCharges'].agg(['mean','median','min','max', 'count'])
print(charge_stats)
```

```python
data['ChurnBinary'] = data['Churn'].map({'Yes':1, 'No':0})
```

```python
churn_stats = data.groupby('TenureGroup')['ChurnBinary'].mean()*100
print(churn_stats)
```

```python
tenure_summary = data.groupby('TenureGroup').agg(
    avg_monthly_charges=('MonthlyCharges', 'mean'),
    median_monthly_charges=('MonthlyCharges', 'median'),
    churn_rate=('ChurnBinary', 'mean'),
    customer_count=('ChurnBinary', 'count')
)

# convert churn rate to %
tenure_summary['churn_rate'] *= 100

print(tenure_summary)
```

```python
gender_churn = data.groupby('gender')['ChurnBinary'].mean()*100
print(gender_churn)
```

```python
senior_churn = data.groupby('SeniorCitizen')['ChurnBinary'].mean()
print(senior_churn)
```

```python
plt.figure(figsize=(5,4))
gender_churn.plot(kind='bar')
plt.title('Churn Rate by Gender')
plt.ylabel('Churn Rate (%)')
plt.show()

plt.figure(figsize=(5,4))
senior_churn.plot(kind='bar')
plt.title('Churn Rate by Senior Citizen Status')
plt.ylabel('Churn Rate (%)')
plt.show()
```

```python
#it shows senior citizens churn more
# gender usually does not strongly affect churn
```

```python
payment_churn = data.groupby('PaymentMethod')['ChurnBinary'].mean()*100
contract_churn = data.groupby('Contract')['ChurnBinary'].mean()*100
print(payment_churn)
print(contract_churn)
```

```python
plt.figure(figsize=(6,4))
contract_churn.plot(kind='bar')
plt.title('Churn Rate by Contract Type')
plt.ylabel('Churn Rate (%)')
plt.show()

plt.figure(figsize=(6,4))
payment_churn.plot(kind='bar')
plt.title('Churn Rate by Payment Method')
plt.ylabel('Churn Rate (%)')
plt.show()
```

```python
# it shows month to month customers churn most
#credit card and bank transfer reduce churn
#electronic check = risk segment
```
